```python
Tiempo = np.array([0.2,0.4,0.6,0.8,1,1.2,1.4,1.6,1.8])
Velocidad= np.array([45,91,168,196,264,315,371,416,497])
dict_1= {"Tiempo":Tiempo,
          "Velocidad":Velocidad
         }
punto2 = pd.DataFrame(dict_1)
```

```python
Volts = [1.00,2.00,3.00,4.00,5.00,6.00,7.00,8.00]
Intensidad = [2.00,2.80,3.45,4.00,4.45,4.90,5.30,5.70]
Calor = [0.15,0.62,1.30,2.20,3.20,4.45,5.90,7.50]
a_dict = {"Volts":Volts,
          "Intensidad":Intensidad,
          "Calor": Calor
         }
punto3 = pd.DataFrame(a_dict)
```

```python
punto3
```

```python
fig = plt.figure(figsize=(15,7))
#Posición de Gráficos
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
#Titulos de Ejes
ax1.set_xlabel('Calor', fontsize=20)
ax1.set_ylabel('Voltaje', fontsize=20)
ax2.set_xlabel('Calor', fontsize=20)
ax2.set_ylabel('Voltaje', fontsize=20)
#Titulos de gráficos
ax1.title.set_text('Voltaje en función de calor (Escala Logaritmica)')
ax2.title.set_text('Voltaje en función de calor')
#Escalas de los gráficos
ax1.set_xscale('log')
ax1.set_yscale('log')
#R2 Coeff
#R2COEFF = 'R2 Coeff es ' + str(punto3['Volts'].corr(punto3['Calor']))
#ax2.text(1, 6,R2COEFF, fontsize=15)
#Plots
ax2.scatter(Calor, Volts)
ax1.scatter(Calor, Volts)
ax2.plot(Calor, Volts,c='r')
ax1.plot(Calor, Volts,c='r')
fig.savefig('TALLER3/PUNTO3/Comp_Calor.png')
```

```python
fig = plt.figure(figsize=(15,7))
#Posición de Gráficos
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
#Titulos de Ejes
ax1.set_xlabel('Intensidad', fontsize=20)
ax1.set_ylabel('Voltaje', fontsize=20)
ax2.set_xlabel('Intensidad', fontsize=20)
ax2.set_ylabel('Voltaje', fontsize=20)
#Titulos de gráficos
ax1.title.set_text('Voltaje en función de intensidad (Escala Logaritmica)')
ax2.title.set_text('Voltaje en función de intensidad')
#Escalas de los gráficos
ax1.set_xscale('log')
ax1.set_yscale('log')
#R2 Coeff
#R2COEFF = 'R2 Coeff es ' + str(punto3['Volts'].corr(punto3['Calor']))
#ax2.text(1, 6,R2COEFF, fontsize=15)
#Plots
ax2.scatter(Intensidad, Volts)
ax1.scatter(Intensidad, Volts)
ax2.plot(Intensidad, Volts,c='r')
ax1.plot(Intensidad, Volts,c='r')
fig.savefig('TALLER3/PUNTO3/Comp_Intensidad.png')
```
